```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x=df[["sqft_living"]]
y=df[["price"]]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape,x_test.shape)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
print("intercept:",model.intercept_)
print("slope (b1):",model.coef_[0])
```

```python
from re import M
from sklearn.metrics import mean_squared_error,r2_score
y_predict=model.predict(x_test)
mse=mean_squared_error(y_test,y_predict)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_predict)
print("mean squared error:",mse)
print("root mean squared error:",rmse)
print("R2 score:",r2)
```

```python
plt.scatter(x_test,y_test,color="blue",alpha=0.5,label="Actual prices")
plt.plot(x_test,y_predict,color="red",linewidth=2,label="Regression line")
plt.xlabel("sqft_living")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
x_multi=df[["sqft_living","bedrooms","bathrooms"]]
y=df["price"]
x_train,x_test,y_train,y_test=train_test_split(x_multi,y,test_size=0.2,random_state=42)

model_multi=LinearRegression()

model_multi.fit(x_train,y_train)

y_predict_multi=model_multi.predict(x_test)
print("Intercept:",model_multi.intercept_)
print("coefficients:",model_multi.coef_)
print("R2 score:",r2_score(y_test,y_predict_multi))
```

```python
top_features=["sqft_living","grade","sqft_above","sqft_living15","bathrooms","view"]
x_top=df[top_features]
x_train,x_test,y_train,y_test=train_test_split(x_top,y,test_size=0.2,random_state=42)
model_top=LinearRegression()
model_top.fit(x_train,y_train)
y_pred_top=model_top.predict(x_test)
print("Intercept:",model_top.intercept_)
print("coefficients:",dict(zip(top_features,model_top.coef_)))
print("R2 score:",r2_score(y_test,y_pred_top))
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif_data=pd.DataFrame()
vif_data["feature"]=x_top.columns

vif_data["VIF"]=[variance_inflation_factor(x_top.values,i) for i in range(len(x_top.columns))]

print(vif_data)
```

```python
selected_features=["sqft_living","grade","view"]
x_sel=df[selected_features]
y=df["price"]
x_train,x_test,y_train,y_test=train_test_split(x_sel,y,test_size=0.2,random_state=42)
model_sel=LinearRegression()
model_sel.fit(x_train,y_train)
y_pred_sel=model_sel.predict(x_test)
print("Intercept:",model_sel.intercept_)
print("coefficent:",dict(zip(selected_features,model_sel.coef_)))
print("R2 score:",r2_score(y_test,y_pred_sel))
```

```python
residuals=y_test-y_pred_sel
plt.figure(figsize=(8,6))
sns.histplot(residuals,kde=True)
plt.title("Residual Distribution")
plt.show()

plt.figure(figsize=(8,6))
plt.scatter(y_pred_sel,residuals,alpha=0.5)
plt.axhline(y=0,color="red",linestyle="--")
plt.xlabel("predicted prices")
plt.ylabel("Residuals")
plt.title("Residual plot")
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model_sel,x_sel,y,cv=5,scoring="r2")
print("cross-validated R2 scores:",scores)
print("mean R2:",np.mean(scores))
```
